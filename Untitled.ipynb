```python
df.columns = df.columns.str.strip()
```

```python
# Fill numerical missing values with mean or mode
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].mean())
df['Engine'] = df['Engine'].fillna(df['Engine'].mean())
df['Power'] = df['Power'].fillna(df['Power'].mean())
df['Seats'] = df['Seats'].fillna(df['Seats'].mode()[0])
```

```python
df.info()
```

```python
#create new feature car age
df['Car_Age'] = 2024 - df['Year']
# Drop `Year` as it's redundant after this transformation
df = df.drop(columns=['Year'])
```

```python
df.head()
```

```python
#Name Simplification
#Extract the brand of the car from the Name column:
df['Brand'] = df['Name'].str.split(' ').str[0]
# Drop `Name` as it's too detailed and may not add value to the analysis
df = df.drop(columns=['Name'])
```

```python
df = pd.get_dummies(df, drop_first=False)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
 #Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['sqrt', 'log2']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestRegressor()
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
rf_random.fit(X_train,y_train)
```
